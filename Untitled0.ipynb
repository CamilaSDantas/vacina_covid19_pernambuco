```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ocupacaoLeitos = pd.read_csv('/content/drive/MyDrive/DADOS PESQUISA VACINA LEITOS/esus-vepi.LeitoOcupacao.csv')
```

```python
ocupacaoLeitos.head()
```

```python
ocupacaoLeitos_mask=ocupacaoLeitos['estado']=='Pernambuco'
```

```python
filtered_ocupacaoLeitos = ocupacaoLeitos[ocupacaoLeitos_mask]
```

```python
colunas = ['ocupacaoConfirmadoUti', 'dataNotificacao']
filtered_ocupacaoLeitos = filtered_ocupacaoLeitos[colunas]
```

```python
filtered_ocupacaoLeitos.head()
```

```python
filtered_ocupacaoLeitos.dataNotificacao = filtered_ocupacaoLeitos.dataNotificacao.astype('datetime64')
```

```python
filtered_ocupacaoLeitos.tail()
```

```python
mask = (filtered_ocupacaoLeitos['dataNotificacao'] > '2020-04-15') & (filtered_ocupacaoLeitos['dataNotificacao'] <= '2022-02-01')
```

```python
uti_ocupacaoLeitos =filtered_ocupacaoLeitos.loc[mask]
```

```python
uti_ocupacaoLeitos.isnull().sum()
```
